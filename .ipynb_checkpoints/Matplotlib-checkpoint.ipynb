```python
#Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.
import matplotlib.pyplot as plt
# %matplotlib inline  
# by writing this we dont need to write plt.show everytime we need to display the graph
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
plt.scatter(x,y,c='r')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("graph in 2d")
plt.savefig("test.png") #save the image
plt.show()
```

```python
y=x*x
```

```python
plt.plot(x,y,'b*',linestyle="dashed",linewidth=2,markersize=12)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("graph in 2d")
plt.show()
```

```python
#creating subplots
plt.subplot(2,2,1) #no. of rows,cols,position
plt.plot(x,y,'b')
plt.show()
plt.subplot(2,2,2) #no. of rows,cols,position
plt.plot(x,y,'g*')
plt.show()
plt.subplot(2,2,3) #no. of rows,cols,position
plt.plot(x,y,'r--')
plt.show()
```

```python
#bar plot
x=[2,8,10]
y=[2,2,11]
x2=[12,8,11]
y2=[23,22,11]
plt.bar(x2,y2,color='g')
plt.bar(x,y,color='r')
plt.show()
```

```python
a=np.array([22,3,4,43,34,43,5,5])
plt.hist(a,bins=20)
plt.show()
```

```python
#box plot helps to find percentiles
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True); #if vert=false then it will be horizontal and if patch_artist=false then colour will go 
plt.show()
```
