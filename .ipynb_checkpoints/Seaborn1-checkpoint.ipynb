```python
# if we analyse a single feature f1 ->univariate analysis
# if we analyse 2 features f1,f2 ->bivariate analysis
# distribution plots -> 1. distplot 2. joinplot 3. pairplot
#for bivariate analysis we need to visualise in 2d diagram so we use joinplot
# for more than two features visualisation we use paiplot
import seaborn as sns
```

```python
df=sns.load_dataset("tips") # this dataset is already there in seaborn ,our task is to find the tip based on the other data
df.head()
```

```python
#tip->dependent feature rest all are independent
# heatmap gives us a correlation between different features and it can only be applied to floating point numbers and not categorical features
# this is bcz categorical features are object based and correlation ranges from -1 to 1
df.dtypes
```

```python
df.corr(numeric_only=True)  # as it wont be able to convert string to float
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

```python
# univariate/bivariate analysis -> jointplot
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')  # hex mean instead of dots we will get plot in form of hexagons
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')  # instead of hex if we give reg it will draw a pdf(probab. density function) and a regression line(best fit line)
```

```python
#pairplot -> also called scatterplot -> more than 2 independent features->it does the permutation and combination of all features and again it takes only int or float values
sns.pairplot(data=df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='smoker')
```

```python
df['smoker'].value_counts()
```

```python
#distplot -> used to make histograms 
sns.distplot(df['tip'])
# it shows histogram and if by default kde(kernel density estimation) is present it shows wrt density
# instead of count onlu % will be mentioned in y axis
```

```python
# if we write kde=false then it shows the count
sns.distplot(df['tip'],kde=False,bins=10)
```

```python
#Categorical plots -> work for categorical features(sex,smoke,day,time)
sns.countplot(x='day',data=df) # counts no. of observations
# give either x/y as other axis will always be count
```

```python
sns.countplot(y='day',data=df)
```

```python
# barplot -> both x and y have to be given unlike countplot
sns.barplot(x='sex',y='tip',data=df)
```

```python
# boxplot -> indicates 0,25,50,100 percentile and dots are outliers
# we dont have categorial features
sns.boxplot(x='sex',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(df,orient='v')  # since x n y are not mentioned it takes all the values
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

```python
#violin plot -> combination of boxplot and kde
# inner part is boxplot and outer part is kde
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
